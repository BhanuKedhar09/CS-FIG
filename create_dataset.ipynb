```python
import json
from datasets import load_dataset
import json
import glob
from datasets import Dataset
import pandas as pd
import numpy as np
```

```python

def normalize_json_data(file_paths):
    all_data = []
    all_keys = set()
    
    for file_path in file_paths:
        with open(file_path, 'r') as file:
            data = json.load(file)
            all_data.extend(data)
            for item in data:
                all_keys.update(item.keys())
    
    normalized_data = []
    for item in all_data:
        normalized_item = {key: item.get(key, None) for key in all_keys}
        normalized_data.append(normalized_item)
    
    return normalized_data

json_files = glob.glob('/lstr/sahara/datalab-ml/z1974769/outputs_pdffigures/real_outputs/*.json')

normalized_data = normalize_json_data(json_files)
```

```python
with open("/lstr/sahara/datalab-ml/z1974769/outputs_pdffigures/real_outputs/all_files_data_combined.json", "w") as f:
    json.dump(normalized_data, f)
    
```
